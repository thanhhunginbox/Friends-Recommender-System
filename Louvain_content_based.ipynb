```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from numpy import dot
from numpy.linalg import norm
```

```python
def cosine(a, b):
    return dot(a, b)/(norm(a)*norm(b))
```

```python
community = pd.read_csv('CrawlData/data/reprocessed_data/louvain_communities/community15.csv')
community
```

```python
relationship = pd.read_csv('CrawlData/data/reprocessed_data/relationship_content_based.csv')
```

```python
data = pd.read_csv('CrawlData/data/cleaned.csv')
data['string']=data[['Workplaces', 'College','School', 'Places','Gender' ]].agg(' '.join, axis=1)
```

```python
ls = []
l = list(community['id'])
for i, r in data.iterrows():
    if r['id'] in l:
        ls.append(r)
data_community = pd.DataFrame(ls)
```

```python
# data_community.reset_index(drop=True, inplace=True)
```

```python
data_community
```

```python
indices = pd.Series(data_community.index,data_community.id)
```

```python
vectorizer = CountVectorizer()
X1 = vectorizer.fit_transform(data_community["string"])
```

```python
vectors = X1.toarray()
```

```python
target_id = 17
target_vector = vectors[indices[target_id]]
```

```python
cosine_map = []
for i, v in zip(list(data_community['id']), vectors):
    cosine_map.append((i, cosine(target_vector, v)))
```

```python
ranked_list = sorted(cosine_map, key=lambda x: x[1], reverse=True)[1:]
```

```python
friends_list = []
for i, r in relationship.iterrows():
    if r['id1'] == target_id:
        friends_list.append(r['id2'])
    elif r['id2'] == target_id:
        friends_list.append(r['id1'])
```

```python
id_ranked_list = [i[0] for i in ranked_list]
for i in id_ranked_list:
    if i in friends_list:
        print(1)
```
