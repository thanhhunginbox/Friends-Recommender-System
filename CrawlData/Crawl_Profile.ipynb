```python
col = ['id', 'Workplaces', 'College', 'School', 'Places', 'Gender']
```

```python
list_intro = []
```

```python
for i, row in df.iterrows():

    url = str(row['url'])
    if ('php' in url):
        url = url + '&sk='
    else:
        url = url + '/'

    # about_work_and_education
    
    txt = []
    txt.append(i)
    browser.get(url+'about_work_and_education')
    sleep(2.5)

    try:
        intro_div = browser.find_element(
            By.XPATH, '/html/body/div[1]/div[1]/div[1]/div/div[3]/div/div/div/div[1]/div[1]/div/div/div[4]/div/div/div/div[1]/div/div/div/div/div[2]/div/div')
    except:
        txt.append('none')
        txt.append('none')
        txt.append('none')

    try:
        html = intro_div.get_attribute('innerHTML')
    except:
        list_intro.append([i, '','','','',''])
        print([i, '','','','',''])
        continue
    soup = BeautifulSoup(html, 'html.parser')

    children = soup.findChildren('div', recursive=False)
    for c in children:
        t = []
        sub_child = c.findChildren('div', recursive=False)
        for sc in sub_child[1:]:
            span = c.findChildren('span')
            t.append(span[1].text.lower())
        txt.append('|'.join(t))

    # about_places

    browser.get(url+'about_places')
    sleep(2.5)

    try:
        intro_div = browser.find_element(
            By.XPATH, '/html/body/div[1]/div/div[1]/div/div[3]/div/div/div/div[1]/div[1]/div/div/div[4]/div/div/div/div[1]/div/div/div/div/div[2]/div/div/div')
    except:
        txt.append('none')

    try:
        html = intro_div.get_attribute('innerHTML')
    except:
        list_intro.append([i, '','','','',''])
        print([i, '','','','',''])
        continue
    soup = BeautifulSoup(html, 'html.parser')
    children = soup.findChildren('div', recursive=False)
    t = []
    for c in children[1:]:
        span = c.findChildren('span')
        t.append(span[0].text.lower())
    txt.append('|'.join(t))

    # # Basic info

    browser.get(url+'about_contact_and_basic_info')
    sleep(2.5)

    try:
        intro_div = browser.find_element(
            By.XPATH, '/html/body/div[1]/div/div[1]/div/div[3]/div/div/div/div[1]/div[1]/div/div/div[4]/div/div/div/div[1]/div/div/div/div/div[2]/div/div/div/div[3]/div[2]/div')
    except:
        txt.append('none')

    try:
        html = intro_div.get_attribute('innerHTML')
    except:
        list_intro.append([i, '','','','',''])
        print([i, '','','','',''])
        continue
    
    soup = BeautifulSoup(html, 'html.parser')
    if 'ale' in soup.text:
        txt.append(soup.text.replace('Gender', '').lower())
    else:
        txt.append('notknown')

    list_intro.append(txt)
    print(txt)
    if(len(list_intro)>=300):
        pd.DataFrame(list_intro, columns=col).to_csv(f'data/info/cc{i}.csv', index=False)
        list_intro = []
pd.DataFrame(list_intro, columns=col).to_csv(f'data/info/cc{i}.csv', index=False)
```
