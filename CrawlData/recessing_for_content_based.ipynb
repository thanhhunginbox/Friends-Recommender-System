```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from wordcloud import WordCloud
from unidecode import unidecode
```

```python
df = pd.read_csv("data/data.csv",encoding = 'utf-8')
```

```python
df.head(5)
```

## Xu ly Workplaces columns 

```python
import re
def Processing_workplaces(string):
    string = unidecode(string)
    cleanString = re.sub('\W+',' ', string.replace("no workplaces to show","No_workplaces").lower().strip())
    return cleanString
```

```python
df['Workplaces']= df['Workplaces'].map(lambda x: Processing_workplaces(x))
```

```python
d_frame = df.Workplaces

plt.subplots(figsize=(10,10))
wordcloud = WordCloud(
                   background_color='White',
                        width = 700,
                        height = 400
    ).generate(" ".join(d_frame))

plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('graph.png')
```

```python
d_frame = df.College

plt.subplots(figsize=(10,10))
wordcloud = WordCloud(
                   background_color='White',
                        width = 700,
                        height = 400
    ).generate(" ".join(d_frame))

plt.imshow(wordcloud)
plt.axis('off')
```

```python
import re

def Processing_college(string):
    replacing_words = ["studied at", "studies", "studies at","studied" ,"at" ]
    for i in replacing_words:
        string = string.replace(i,'')
    cleanString = re.sub('\W+' ,' ', unidecode(string.replace("no schools to show","no_College").lower().strip()))
    
    return cleanString
```

```python
df.College=df.College.map(lambda x: Processing_college(x))
```

```python
def Processing_School(string):
    replacing_words = ["went to", "goes to","at" ]
    for i in replacing_words:
        string = string.replace(i,'')
    cleanString = re.sub('\W+' ,' ', unidecode(string.replace("no schools to show","no_School").lower().strip()))
    
    return cleanString
```

```python
df.School=df.School.map(lambda x: Processing_School(x))
```
